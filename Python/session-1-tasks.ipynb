### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# Write your code here
# Input temperature in Celsius
celsius = float(input("Enter temperature in Celsius: "))

# Convert Celsius to Fahrenheit
fahrenheit = (celsius * 9/5) + 32

# Display the result
print(f"{celsius} degrees Celsius is equal to {fahrenheit} degrees Fahrenheit")
````

### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
# Write your code here
# taking two integer input from the user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
print("Number before swapping: ", num1, num2)

temp = num1
num1 = num2
num2 = temp

print("Number after swapping: ", num1, num2)
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
# Importing math library for sqrt() function
import math

# Input coordinates for the first point
x1 = float(input("Enter x coordinate of the first point: "))
y1 = float(input("Enter y coordinate of the first point: "))

# Input coordinates for the second point
x2 = float(input("Enter x coordinate of the second point: "))
y2 = float(input("Enter y coordinate of the second point: "))

# Calculate the Euclidean distance
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Display the result
print(f"The Euclidean distance between the points is: {distance}")
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.



````python
# Write your code here
# Input values from the user
principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the rate of interest (in %): "))
time = float(input("Enter the time period (in years): "))

# Calculate the simple interest
simple_interest = (principal * rate * time) / 100

# Display the result
print(f"The simple interest is: {simple_interest}")
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Get the total number of heads and legs from the user
heads = int(input("Enter the total number of heads: "))
legs = int(input("Enter the total number of legs: "))

# Error checking: the number of legs should be even and at least twice the number of heads
if legs % 2 != 0 or heads > legs or legs > 4 * heads:
    print("Invalid input")
else:
    # Calculate the number of dogs
    dogs = (legs - 2 * heads) // 2

    # Calculate the number of chickens
    chickens = heads - dogs

    print(f"There are {chickens} chickens and {dogs} dogs.")

````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
num = int(input("Enter the number: "))
total_sum = 0

for n in range(num+1):
  total_sum += n**2


print(total_sum)
````


### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Get user inputs for the first two terms, common difference, and n
a1 = float(input("Enter the first term (a1) of the arithmetic series: "))
a2 = float(input("Enter the second term (a2) of the arithmetic series: "))
n = int(input("Enter the value of n for which you want to find the nth term: "))

# here d is the difference between the
d = a2 - a1

# Calculate the nth term using the formula
nth_term = a1 + (n - 1) * d

print(f"The {n}th term of the arithmetic series is: {nth_term}")
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
numerator_1 = float(input("Enter the value of numerator1 of a fraction: "))
denominator_1 = float(input("Enter the value of denominator1 of a fraction: "))

numerator_2 = float(input("Enter the value of numerator2 of a fraction: "))
denominator_2 = float(input("Enter the value of denominator2 of a fraction: "))

t_n = numerator_1*denominator_2 + numerator_2*denominator_1
t_d = denominator_1 * denominator_2

sum = t_n / t_d

print("Sum of the two fraction is {}".format(sum))
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# milk tank acts as a cubiod
# glass acts as a sphere
import math

mt_h = float(input("Enter the height of milk tank: "))
mt_l = float(input("Enter the length of milk tank: "))
mt_b = float(input("Enter the breadth of milk tank: "))

g_h = float(input("Enter the height of glass: "))
g_r = float(input("Enter the radius of glass: "))

mt_v = mt_l * mt_b * mt_h
g_v = math.pi * g_r * g_r * g_h

print("Number of glasses is: ", math.floor(mt_v / g_v))
````
